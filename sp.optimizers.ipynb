### optimizers

```python
from scipy.optimize import minimize_scalar

minimize_scalar(lambda x: (x - 2) ** 2), \
minimize_scalar(lambda x: x**2 + x + 2)
```

```python
from scipy.optimize import minimize

minimize(lambda x: x**2 + x + 2, 0, method='BFGS')
```

```python
from scipy.optimize import root
from math import cos

#works better non-linear problems

root(lambda x: (x-1) ** 3, 0), \
root(lambda x: x + cos(x), 0)
```
