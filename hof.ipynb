## higher order functions

```python
#sorting
#sort() vs sorted()

names = ['John', 'Alice', 'Bob', 'Diana']

#names.sort() # in-place sort
#names.sort(key=lambda x: len(x))   #in-place sort

names_sorted = sorted(names) 
names_sorted_by_len = sorted(names, key=lambda x: len(x))  # sort by length of names


names_sorted, names_sorted_by_len
```

```python
#enumarate = indexing
# always convert to list, tuple, set or dict ...

enumerate_names = list(enumerate(names))

enumerate_names
```

```python
#map
# compute w/ lambda
# convert convert convert to list tuple dict set...
# in your mind decide what to convert to ... a new list, tuple, set or dict ...
names_length = map(lambda x: len(x), names) 

# tuple(zip(names_length, names))
# dict(zip(names_length, names))
list(zip(names_length, names))
```

```python
#filter
# returns true/false based on lambda condition
names_len_greater_then_3 = filter(lambda e: len(e) > 3, names)

list(names_len_greater_then_3)
```

```python
#reduce
from functools import reduce    
names_length = map(lambda x: len(x), names) 

#reduce requires 2 paramaters
reduce(lambda x,y: x+y, list(names_length), 0)
```

```python
names_length = map(lambda x: len(x), names) 

reduce(lambda x,y: x+y, 
       set(
           map(lambda x: x ** 2, 
               filter(lambda e: e % 2 != 0, names_length))))
```

```python
# min, max, sum
#below, min is used with map 
#so, it aint the built-in min function

def debug_min(x):
    print(f"debug_min called with {x} {min(x, key=lambda e: x[2:])}")
    return min(x)  # min by second letter

names = ['John', 'Alice', 'Bob', 'Diana']
names_min = map(lambda k: debug_min(k), names)

list(names_min)
```

```python
# min, max, sum
# ord('o')  # 111 default comparision is lexicographical

grades = [88, 92, 79, 93, 85]
names = ['John', 'Alice', 'Bob', 'Diana']

mm = min(names, key=lambda x: x[1].lower())  # min by second letter
xx = max(grades, key=lambda x: x % 10)  # max by last digit

mm, xx
```

```python
# zip 
names = ["ali", "veli", "deli", "selim"]
grades = [3.2, 4.5, 5.0, 2.5]

for name, grade in zip(names, grades):
    print(f"{name}: {grade}")
```
