### sparse

```python
import scipy.sparse as sp
import numpy as np

sparse2D = np.array([[0, 0, 1], [1, 0, 0], [0, 0, 2]])
print(sparse2D) #3x3 [][]

#csrow - based
matrix = sp.csr_matrix(sparse2D)
print(matrix)    # (row=0, column=2)	-> 1
```

```python
#values
matrix.data    # array([1, 1, 2])
```

```python
matrix.count_nonzero  # 3 out of 9 elements
```

```python
matrix_csc = matrix.tocsc()

print(matrix_csc) 
```
