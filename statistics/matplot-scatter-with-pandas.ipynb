```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
    "name": ["ali", "veli", "deli", "helli", "melli"],
    "age": [20, 30, 22, 25, 31],
    "gender": ["M", "M", "F", "F", "M"],
    "state": ["ca", "ca", "dc", "tx", "ca"],
    "numberOfChildren": [3, 2, 1, 4, 5],
    "numOfPets": [1,0,0,1,2]
})
```

## 1 st way

```python
# pandas 2 plot
df.plot(x="age", y="numOfPets")
```

## 2nd way

```python

x = df["age"].tolist()
y = df["numOfPets"].tolist()
x, y
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y)
ax.grid(True)
plt.show()
```
