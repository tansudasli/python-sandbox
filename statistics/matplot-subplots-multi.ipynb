```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(low=0, high=10, size=50)
y = np.random.uniform(low=7, high=10, size=50)
x, y
```

```python
fig, ax = plt.subplots(2, 1)
```

```python
# plot 1
plt.subplot(2,1,1)
plt.plot(x, y, "o")

plt.title("2-subplots")
plt.ylabel("Test-1 score")

# plot 2
plt.subplot(2,1,2)
plt.plot(y, x, ".")

plt.xlabel("Student number")
plt.ylabel("Test-2 score")

#
plt.show()
plt.savefig("matplot-subplots-multi.png")
```
