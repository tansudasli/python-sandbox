### hof and iteration

reframe problems as higher orde functions, then iterate if needed

- in hof, no need to loop or for..in
- you only need for..in when to print the results (mapped, reduced etc.) if dict, nested ...

and, you should make a bridge between python, numpy and pandas. cause, hof are more concise in pandas...

```python
# classic iteration for ..in ..:#
even = [0,2,4,6,8,10,1,14,16,18,20]

# no need to loop, not dict or nested structure !
# for n in even:
#     print(n, end=' ')

print(even)
```

```python
# think problem as higher order functions
# count how many numbers are less than 20 in a list 
# is filtering + counting + printing

numbers = [1,29,49,20,18,5,4,67,12]

filtered = filter(lambda x: x < 20, numbers)
list(filtered)  # [1, 18, 5, 4, 12]

#classical way
# def countLessThan20(l):
#     counter = 0
#     for i in l:
#         counter += 1 if i < 20 else 0
#     return counter


# countLessThan20(list)

#using collections module
# from collections import Counter
# c = Counter( i < 20 for i in list)  #Counter({True: 5, False: 4})

# c[True]  # count of items less than 20  
```

```python
#list to pandas series, 
# then filter and count
# much better handling with hof in pandas !!
import pandas as pd

s = pd.Series(numbers)

s_filtered = s[s < 20]

s_filtered.tolist(), s_filtered.count()
```
