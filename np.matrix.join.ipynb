`````python
import numpy as np
`````

`````python
# vector 1d

vector_a = np.array([1, 2, 3]) # vector, shape (3,), 1D
vector_b = np.array([4, 5, 6])

concatenated_vector = np.concatenate((vector_a, vector_b)) #axis=0 default
print(f"Concatenated Vector: {concatenated_vector} - Shape: {concatenated_vector.shape}")
`````

`````python
# tensor 2d
matrix_a = np.array([[1, 2, 3], [4, 5, 6]])  # shape (2, 3)
matrix_b = np.array([[7, 8, 9], [10, 11, 12]])  # shape (2, 3)

concatenated_vector_2d_row = np.concatenate((matrix_a, matrix_b), axis=0)  # shape (4, 3)
concatenated_vector_2d_col = np.concatenate((matrix_a, matrix_b), axis=1)  # shape (2, 6)

print(f"Concatenated 2D Matrix along rows (axis=0):\n{concatenated_vector_2d_row}\nShape: {concatenated_vector_2d_row.shape}")
print(f"Concatenated 2D Matrix along columns (axis=1):\n{concatenated_vector_2d_col}\nShape: {concatenated_vector_2d_col.shape}")
`````

`````python
#tensor 3d
tensor_a = np.array([[[1, 2], [3, 4]], 
                     [[5, 6], [7, 8]]])  # shape (2, 2, 2)
tensor_b = np.array([[[9, 10], [11, 12]],
                     [[13, 14], [15, 16]]])  # shape (2, 2, 2)

concatenated_tensor_depth = np.concatenate((tensor_a, tensor_b), axis=2)  # shape (2, 2, 4)
concatenated_vector_3d_row = np.concatenate((tensor_a, tensor_b), axis=0)  # shape (4, 2, 2)
concatenated_vector_3d_col = np.concatenate((tensor_a, tensor_b), axis=1)  # shape (2, 4, 2)

print(f"Concatenated 3D Tensor along depth (axis=2):\n{concatenated_tensor_depth}\nShape: {concatenated_tensor_depth.shape}")
print(f"Concatenated 3D Tensor along rows (axis=0):\n{concatenated_vector_3d_row}\nShape: {concatenated_vector_3d_row.shape}")
print(f"Concatenated 3D Tensor along columns (axis=1):\n{concatenated_vector_3d_col}\nShape: {concatenated_vector_3d_col.shape}")
`````
