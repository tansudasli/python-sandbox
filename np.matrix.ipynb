### numpy

- two differeent way to access [, , ,], or [][][] 
- with some caution
- the other critical thing is shape after slicing !!!

```python
import numpy as np
```

```python
oneD = np.array([1, 2, 3])

#oneD[] not valid

oneD[:], oneD
```

```python
twoD = np.array([[1, 2, 3, 5], [4, 5, 6, 6], [7, 8, 9, 33]])

#twoD[,] not valid
#row, column

twoD[:, :], twoD[:, :].shape, \
twoD[0][2], twoD[0][2].shape, \
twoD[1], twoD[1].shape
```
