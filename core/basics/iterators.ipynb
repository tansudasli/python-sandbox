## iterators

```python
colors = ["red", "green", "blue", "yellow", "black"]

names = ["alice", "bob", "charlie", "david", "eve"]

grades = {"alice": [85, 90, 78], "bob": [70, 75, 80], "charlie": [95, 100, 92]}
```

```python
# enumarate = adding index
# always convert to list, tuple, set or dict ...

list(enumerate(colors)) # [(0, 'red'), (1, 'green'), (2, 'blue'), (3, 'yellow'), (4, 'black')]
```

```python
# map
# compute w/ lambda
# convert convert convert to list tuple dict set...
# in your mind decide what to convert to ... a new list, tuple, set or dict ...

computeLength = map(lambda color: len(color), colors)

#list(computeLength) #[3, 5, 4, 6, 5]
#dict(computeLength, names) #{'alice': 5, 'bob': 3, 'charlie': 7, 'david': 5, 'eve': 3}
#dict(zip(computeLength, names)) #{'alice': 5, 'bob': 3, 'charlie': 7, 'david': 5, 'eve': 3}
tuple(zip(computeLength, names)) #(3, 5, 4, 6, 5)
```

```python
colorsNamesMapped = map(lambda a, b: a + " " + b, names, colors )

# list(colorsNamesMapped) #['alice red', 'bob green', 'charlie blue', 'david yellow', 'eve black']

for k, v in dict(map(lambda a, b: (a, b), colors, names)).items():
    print(f'{k} : {v}')
```

```python
#filter

filteredColors = filter(lambda color: 'e' in color, colors) # ['red', 'green', 'blue', 'yellow', 'black']
filteredGrades = filter(lambda grade: sum(grade)/len(grade) >= 80, grades.values()) # [[85, 90, 78], [70, 75, 80], [95, 100, 92]]

list(filteredColors), list(filteredGrades) 
```
