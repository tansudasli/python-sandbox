```python
grades = { 
    'Ali' : [10, 80, 90], 
    'Veli': [80, 80, 90] 
    }


#accessing dictionary elements

# grades.keys()
# grades.values()
# grades.items()    #dict_items object. different than list of tuples !!!

#k, v unpacking is possible, if .items() used !!!
for k, v in grades.items():
    print(f'{k}: {v}')

#grades.get('Ali')
#grades['Ali']
#grades['Veli'][1]
```

```python
#comprehension 
#version 2
average_of_grades = { k : sum(v)/len(v) for k, v in grades.items() }

# comprehension = writing in 0ne-line 
#  { 
#   k : sum(v)/len(v) 
#   for k, v in grades.items() 
#   }

#version 1
# for k, v in grades.items():
#     print(f'{k}: {sum(v)/len(v)}')


average_of_grades
```

```python
# one step futher
# iterators with map, zip
# comperhension with map + lambda
# lambda
# higher order functions

compute_average_of_grades = map(lambda sum, len: sum/len,          #lambda
                                [sum(v) for v in grades.values()], #iterator w/ comprehension
                                [len(v) for v in grades.values()])



#what we get from map
# list(compute_average_of_grades)  #[60.0, 83.33333333333333]

#blend w/ zip instead of for-loop
#always combines as pairs !
dict(zip(grades.keys(), compute_average_of_grades)) #{'Ali': 60.0, 'Veli': 83.33333333333333}
```
