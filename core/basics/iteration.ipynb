## Iterating over Dictionaries

In this exercise you will use the same dictionaries as the ones we used in the lesson - "prices" and "quantity". This time, don't just calculate all the money Jan spent. Calculate how much she spent on products with a price of 5 dollars or more.

```python
prices = {"box_of_spaghetti" : 4,
          "lasagna"  : 5,
          "hamburger" : 2
        }

quantity = {"box_of_spaghetti" : 6,
            "lasagna"  : 10,
            "hamburger" : 0
        }

prices
```

And how much did Jan spent on products that cost less than 5 dollars?

```python
money_spent_greaater_than_5 = 0
money_spent_less_than_5 = 0

#tenary operators
for item in prices:
    money_spent_greaater_than_5 += prices[item] * quantity[item] if prices[item] >= 5 else 0
    money_spent_less_than_5 += prices[item] * quantity[item] if prices[item] < 5 else 0


# for item in prices:
#     if prices[item] >= 5:
#         money_spent += prices[item] * quantity[item]
#     else:
#         money_spent_less_than_5 += prices[item] * quantity[item]


# for i in quantity:
#     if prices[i] >= 5:
#         money_spent += prices[i] * quantity[i]
money_spent = money_spent_greaater_than_5 + money_spent_less_than_5

money_spent_greaater_than_5, money_spent_less_than_5, money_spent
```
