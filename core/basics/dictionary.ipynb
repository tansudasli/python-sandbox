Create a new dictionary that contains the first five meals as keys and assign the following five values as prices (in dollars):
10, 5, 8, 12, 5. 
Start by *Price_list = {}*.

```python
Prices = {}
values = [100.0, 50.0, 8, 12]
index = 0

for k in Menu:
    print(Menu[k])
    Prices[Menu[k]] = values[index]
    index = index + 1


#Prices = {Menu['meal_1']:10,
#          Menu['meal_2']:5,
#          Menu['meal_3']:8,
#          Menu['meal_4']:12,
#          Menu['meal_5']:5}

Prices
```

Use the *.get()* method to check the price of the Spaghetti.

```python
value = Prices.get('Spaghetti')
print(value)

# 1 - use for to iterate dict items
for key in Prices.items():
    print(key)

#2 - convert dict_items to list or dict!!!
print(Prices.items()) # dict_items object .. a different type
#print(list(Prices.items())[0])
print(
    dict(Prices.items())["Spaghetti"]
    )

print(Prices.values())
list(Prices.values())[0]
```
