## generators

```python
# lazy load
numbers = [1, 2, 3, 4, 5, 6, 7]

# (.. for .. in .. if ..) is a kind of generator expression 
for v in (z ** 2 for z in numbers if z % 2 != 0):
    print(v, end=' ')
```

```python
# generator = define a function, return w/ yield, access with next()
# it is an object, 
# lazy loading
# instead of [], we used () !!!

genarator = (z ** 2 for z in numbers if z % 2 != 0)

genarator, next(genarator), next(genarator)
```

```python
#
not_a_generator_like_above = [z ** 2 for z in numbers if z % 2 != 0]

not_a_generator_like_above != type(genarator)
```

```python
#generators
#uses yield instead of return
#saves states, generates when called next()
def squared_generator(nums):
    for z in nums:
        if z % 2 != 0:
            yield z ** 2

            

gen = squared_generator(numbers) #creates a generator object, then use it

next(gen), next(gen), next(gen)
```

```python
#generators
#uses yield instead of return
#saves states, generates when called next()
def generate_numbers(n):
    for i in range(n):
        yield i * 2 #return = yield + pauses

gen = generate_numbers(40) #creates a generator object, then use it

gen, gen.__next__(), gen.__next__(),gen.__next__()
```
