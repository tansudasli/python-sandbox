```python
#comprehension = one-line for loop

#i, j, k
#threshold = max sum
outlet, item, revenue = 3, 3, 3
threshold = 3

matrix = [(i, j, k) for i in range(outlet)
                    for j in range(item)
                    for k in range(revenue) if (i + j + k) < threshold]

# ar = []
# for i in range(x+1):
#     for j in range(y+1):
#         for k in range(z+1):
#             if ((i+j+k) != n):
#                 ar.append([i, j, k])

matrix
```
