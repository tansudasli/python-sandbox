## iterators

```python
colors = ["red", "green", "blue", "yellow", "black"]
names = ["alice", "bob", "charlie", "david", "eve"]
```

```python
#enumarate, a kind of iterator
colorsIndexed = enumerate(colors)

#colorsIndexed #object
# list(colorsIndexed)[0]  #(0, 'red')

list(colorsIndexed) 
```

```python
#map
#convert convert convert to list tuple dict set...
colorsMapped = map(lambda color: len(color), colors)

list(colorsMapped) #[3, 5, 4, 6, 5]
dict(map(lambda name: (name, len(name)), names)) #{'alice': 5, 'bob': 3, 'charlie': 7, 'david': 5, 'eve': 3}
```

```python
colorsNamesMapped = map(lambda a, b: a + " " + b, names, colors )

# list(colorsNamesMapped) #['alice red', 'bob green', 'charlie blue', 'david yellow', 'eve black']

for k, v in dict(map(lambda a, b: (a, b), colors, names)).items():
    print(f'{k} : {v}')
```
