```python
#groupby
from itertools import groupby

#         item     type
data = [("apple",  "fruit"), 
        ("carrot",  "vegetable"), 
        ("banana",  "fruit"),
        ("spinach", "vegetable")] 

grouped_type = groupby(data, lambda x: x[1])

# list(grouped_type) #object of (key, v=group_iterator) #[('fruit', <itertools._grouper at 0x1111dd8d0>), ..]

#[('fruit', [('apple', 'fruit')])] -> k, v[0]=('apple', 'fruit') !!
# pay attention to the nested structure of groupby 

#list((key, list(group)) for key, group in grouped_type)
#list((key, len(list(group)[0])) for key, group in grouped_type)
```
